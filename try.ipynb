```python
import interface as it
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = it.SimulationBox()
x.set_mu(-3.05)
x
```

```python
rho = np.ones((60, 60, 60))
phi = np.zeros((60, 60, 60))
for i in range(60):
    for j in range(60):
        for k in range(60):
            if (i-29)**2 + (j-29)**2 + (k-29)**2 < 15**2:
                rho[i, j, k] = 0

x.set_rho(rho)
x.set_phi(phi)
x
```

```python
do = it.iterator(x, 10000, 20000)
```

```python
fig, ax = plt.subplots()
ax.plot(range(10000), do.k_history)
plt.xlim(0, 10000)
plt.show()
```
